```python
f = open("9.txt", "r")
lines = f.read().splitlines()
```

```python
def move(hx, hy, tx, ty):
    if abs(hx-tx) > 1:
        tx += (-1, 1)[hx>tx]
        ty += 0 if abs(hy-ty) == 0 else (-1, 1)[hy>ty]
    elif abs(hy-ty) > 1:
        ty += (-1, 1)[hy>ty]
        tx += 0 if abs(hx-tx) == 0 else (-1, 1)[hx>tx]
    return tx, ty

move(3,2,1,1)
```

```python
visit = set()
hx=hy=tx=ty=0
rope = [(0, 0)] * 10
for line in lines:
    dir,n = line.split()
    n = int(n)
    for v in range(int(n)):
        x, y = rope[0]
        match dir:
            case 'R':
                    hx += 1
                    x += 1
            case 'L':
                    hx -= 1
                    x -= 1
            case 'U':
                    hy -= 1
                    y -= 1
            case 'D':
                    hy += 1
                    y += 1
        rope[0] = (x, y)
        visit.add(rope[9])
        tx, ty = move(hx, hy, tx, ty)
        # simulate for 10 knots
        for i, [h, t] in enumerate(zip(rope, rope[1:])):
            rope[i+1] = move(*h, *t)
        print(rope)

visit.add(rope[9])
print(len(visit))
        
    
```
