```python
f = open("12.txt", "r")
lines = f.read().splitlines()
```

```python
from pprint import pprint
d = [list(l) for l in lines]
S = None
E = None
for i, rows in enumerate(d):
    for j, x in enumerate(rows):
        if x == 'S':
            S = (i, j)
            d[i][j] = 'a'
        elif x == 'E':
            E = (i, j)
            d[i][j] = 'z'

print(S, E)
pprint(d)
```

```python
def next_moves(x, y):
    ret = [(x-1, y), (x+1, y), (x, y-1), (x, y+1)]
    return set(e for e in ret if 0 <= e[0] < len(d) and 0 <= e[1] < len(d[0]) and ord(d[x][y]) - ord(d[e[0]][e[1]])  <= 1)
    
print(next_moves(0, 0))
print(next_moves(4, 7))
print(next_moves(0, 7))
print(next_moves(4, 0))
print(next_moves(0, 1))
```

```python
def ganm(prev):
    tmp = set()
    for e in prev:
        tmp = tmp.union(next_moves(*e))
    return tmp
```

```python
ganm({S})
```

```python
spt = {S}
found = False
nxt = {S}
dst = 0
while not found:
    nxt = ganm(nxt)
    for n in nxt:
        spt &= {n}
        if n == E:
            found = True
            break
    dst += 1
print(dst)
```

```python
spt = {E}
found = False
nxt = {E}
dst = 0
while not found:
    nxt = ganm(nxt)
    for n in nxt:
        spt &= {n}
        if d[n[0]][n[1]] == 'a':
            found = True
            break
    dst += 1
print(dst)
```
